```python
import pandas as pd

df = pd.read_csv('/kaggle/input/ice-cream-sales-dataset/Ice Cream.csv')
display(df.head(5))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Cria o gráfico de dispersão
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Temperature', y='Revenue', data=df)

# Adiciona títulos e rótulos
plt.title('Gráfico de Dispersão: Venda vs Temperatura')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Venda')

# Exibe o gráfico
plt.grid(True)
plt.show()
```

```python
import statsmodels.api as sm

# Define as variáveis dependente e independente
y = df['Revenue']  # Variável dependente (Receita)
x = df['Temperature']  # Variável independente (Temperatura)

# Adiciona uma constante ao modelo (para o intercepto)
x = sm.add_constant(x)

# Cria e ajusta o modelo de regressão linear
model = sm.OLS(y, x).fit()

# Exibe o resumo do modelo
print(model.summary())
```
